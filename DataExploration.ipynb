```python
#number of rows with loser_id = winner_id present in tennis.csv 
duplicates = tennis[tennis['winner_id'] == tennis['loser_id']]['winner_id'].values
len(duplicates)
```

```python
duplicate = set()
for index, row in tennis.iterrows():
    if row['winner_id'] in duplicates:
        duplicate.add(row['winner_id'])
    elif row['loser_id'] in duplicates: 
        duplicate.add(row['loser_id'])
        
len(duplicate)
#scanning through tennis.csv for the id of the players that we found playing with themselves, 
#we see that there are only 5 players with this id, so we assume that these rows were wrongly inserted, 
#so we delete them.
```
